```python
import bovespa
```

```python
import pandas as pd
```

```python
from pandas import DataFrame
```

```python
bf = bovespa.File('cotações/COTAHIST_A2012.TXT')
```

```python
regs= []
for ano in range(2008,2020):
    bf = bovespa.File('cotações/COTAHIST_A{0}.TXT'.format(ano))
    for rec in bf.query():
        regs.append(dict([('date', rec.date), ('stock_code', rec.stock_code), ('market', rec.market), 
                                        ('isin_code', rec.isin_code), ('especification', rec.especification), 
                                        ('price_open', rec.price_open), ('price_high', rec.price_high), 
                                        ('price_low', rec.price_low), ('price_mean', rec.price_mean),  
                                        ('price_close', rec.price_close), ('volume', rec.volume),
                                        ('board_lot', rec.board_lot)]))
```

```python
df = DataFrame(regs)
```
